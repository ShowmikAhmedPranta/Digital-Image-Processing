```python
ref = cv2.imread('reference.png',-1)
align = cv2.imread('input.png',-1)

cv2_imshow(align)

#cv2.cvtColor() method is used to convert an image from one color space to another. 

#COLOR_BGR2GRAY    grey color space conversion
img1 = cv2.cvtColor(align, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(ref, cv2.COLOR_BGR2GRAY)
height, width = img2.shape

#creating an object of ORB() function to detect the key points in the image  
#key points mean corner points
orb_detector = cv2.ORB_create(5000)

#ORB_object.detect() function to detect the key points from a given image.
#ORB_object.compute() function to compute the descriptors for a given image.

kp1, d1 = orb_detector.detectAndCompute(img1, None)
kp2, d2 = orb_detector.detectAndCompute(img2, None)

#BFMatcher:
#   Brute-Force matcher is simple. It takes the descriptor of one feature 
#      in first set and is matched with all other features in second set using 
#        some distance calculation. And the closest one is returned.

matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

# Matcher.match() method to get the best matches in two images.
matches = matcher.match(d1, d2)

#We sort them in ascending order of their distances so that best matches
# (with low distance) come to front. Then we draw only first 10 matches 
sorted(matches, key=lambda x: x.distance)
matches = matches[:int(len(matches)*0.9)]
no_of_matches = len(matches)

#2D array object
p1 = np.zeros((no_of_matches, 2))
p2 = np.zeros((no_of_matches, 2))

for i in range(len(matches)):
  p1[i, :] = kp1[matches[i].queryIdx].pt
  p2[i, :] = kp2[matches[i].trainIdx].pt

#Homography Transform: perspective tranformation

homography, mask = cv2.findHomography(p1, p2, cv2.RANSAC)

#warpPerspective:   Fitting the new photo as the size of original photo.
transformed_img = cv2.warpPerspective(align,
                    homography, (width, height))

cv2_imshow(transformed_img)
#print(ref)
```
